# Cryptocurrencies Project: Reducing Data Dimensions Using PCA

In this section we use PCA algorithm from sklearn to reduce the dimensions of the X DataFram down to three principal components.
And after dimensions reduction, a X_pca dataframe is created tha includes all three priciple components value. This dataframe will be joined with another dataframe in the next

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import hvplot.pandas
```

```python
file_path="Resources/X.csv"
X=pd.read_csv(file_path)
X=X.set_index(["Unnamed: 0"])
X.head()
```

```python
X_scaled=StandardScaler().fit_transform(X)
print(X_scaled[0])
X_scaled.shape
```

```python
pca=PCA(n_components=3)
```

```python
X_pca=pca.fit_transform(X_scaled)
```

```python
X_pca=pd.DataFrame(data=X_pca,columns=["PC 1", "PC 2", "PC3"])
X_pca.head(10)
```

```python
file_path="Resources/X_pca.csv"
X_pca.to_csv(file_path,index=False)
```
